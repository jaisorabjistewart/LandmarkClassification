### Why We're Here

Photo sharing and photo storage services like to have location data for each photo that is uploaded. With the location data, these services can build advanced features, such as automatic suggestion of relevant tags or automatic photo organization, which help provide a compelling user experience. Although a photo's location can often be obtained by looking at the photo's metadata, many photos uploaded to these services will not have location metadata available. This can happen when, for example, the camera capturing the picture does not have GPS or if a photo's metadata is scrubbed due to privacy concerns.

If no location metadata for an image is available, one way to infer the location is to detect and classify a discernable landmark in the image. Given the large number of landmarks across the world and the immense volume of images that are uploaded to photo sharing services, using human judgement to classify these landmarks would not be feasible.

````python
# define the CNN architecture
class Net(nn.Module):
    def __init__(self, use_batch_norm=False):
        super(Net, self).__init__()
        
        self.use_batch_norm = use_batch_norm
        n_features = 16
        
        if use_batch_norm:
            # Input: 224x224x3, Kernel: 3x3, Stride: 1, Padding: 1. So Output: (224-3+2)/1 + 1 = 224
            self.conv1 = nn.Conv2d(3, n_features, 3, padding=1, bias=False)
            self.batch_norm1 = nn.BatchNorm2d(n_features)
            
            # 112 x 112 x 16 input tensor (due to pooling layer between conv1 & conv2)
            self.conv2 = nn.Conv2d(n_features, n_features*2, 3, padding=1, bias=False)
            self.batch_norm2 = nn.BatchNorm2d(n_features*2)
            
            # 56 x 56 x 32 input
            self.conv3 = nn.Conv2d(n_features*2, n_features*4, 3, padding=1, bias=False)
            self.batch_norm3 = nn.BatchNorm2d(n_features*4)
            
            # 28 x 28 x 64 input
            self.conv4 = nn.Conv2d(n_features*4, n_features*8, 3, padding=1, bias=False)
            self.batch_norm4 = nn.BatchNorm2d(n_features*8)
            
            # 14 x 14 x 128 input
            self.conv5 = nn.Conv2d(n_features*8, n_features*16, 3, padding=1, bias=False)
            self.batch_norm5 = nn.BatchNorm2d(n_features*16)
            
            # 7 x 7 x 256 input
            #self.conv6 = nn.Conv2d(n_features*16, n_features*32, 3, padding=1, bias=False)
            #self.batch_norm6 = nn.BatchNorm2d(n_features*32)
            
        else:
            self.conv1 = nn.Conv2d(3, n_features, 3, padding=1)
            self.conv2 = nn.Conv2d(n_features, n_features*2, 3, padding=1)
            self.conv3 = nn.Conv2d(n_features*2, n_features*4, 3, padding=1)
            self.conv4 = nn.Conv2d(n_features*4, n_features*8, 3, padding=1)
            self.conv5 = nn.Conv2d(n_features*8, n_features*16, 3, padding=1)
            
        # 7 x 7 x 256 input
        self.conv6 = nn.Conv2d(n_features*16, n_features*32, 3, padding=1)
        # 8 x 8 x 512 input
        ##self.conv7 = nn.Conv2d(512, 1024, 3, padding=1)

        
        # max pooling layer (halves image tensor dimensions)
        self.pool = nn.MaxPool2d(2, 2)
        # linear layer (64 * 4 * 4 -> 500)
        self.fc1 = nn.Linear(512*7*7, 500)
        # linear layer (500 -> 50, since there are 50 possible landmarks)
        self.fc2 = nn.Linear(500, 50)
        # dropout layer (p=0.25)
        self.dropout = nn.Dropout(0.25)
        

    def forward(self, x):
        # add sequence of convolutional and max pooling layers
        x = self.pool(F.relu(self.conv1(x)))
        if self.use_batch_norm:
            x = self.batch_norm1(x)
        x = self.pool(F.relu(self.conv2(x)))
        if self.use_batch_norm:
            x = self.batch_norm2(x)
        x = self.pool(F.relu(self.conv3(x)))
        if self.use_batch_norm:
            x = self.batch_norm3(x)
        x = self.pool(F.relu(self.conv4(x)))
        if self.use_batch_norm:
            x = self.batch_norm4(x)
        x = self.pool(F.relu(self.conv5(x)))
        if self.use_batch_norm:
            x = self.batch_norm5(x)
        x = F.relu(self.conv6(x))
        # flatten image input
        x = x.view(x.size(0), -1)
        # add dropout layer
        x = self.dropout(x)
        # add 1st hidden layer, with relu activation function
        x = F.relu(self.fc1(x))
        # add dropout layer
        x = self.dropout(x)
        # add 2nd hidden layer, without relu activation function
        x = self.fc2(x)
        return x

#-#-# Do NOT modify the code below this line. #-#-#

# instantiate the CNN
model_scratch = Net(use_batch_norm=batch_norm)
print(model_scratch)

# move tensors to GPU if CUDA is available
if use_cuda:
    model_scratch.cuda()
    
````

__Answer:__  I chose to use a small kernel for my convolutions so as to get a lot of depth by the end of the convolutional section. This should allow my model to distinguish more features for a more accurate classification model. By themselves, these convolutional layers wouldn't reduce the size of each layer, (since Input size of images is 224x224x3, Kernel size is 3x3, Stride is 1 and Padding is 1; So Output size is (224-3+2)/1 + 1 = 224. So I used a maxpooling layer between each convolutional layer in order to reduce the dimensionality so as to allow the network to learn features. I then used 2 fully connected layers to go from the many-layered convolutional representation to a prediction of one of the 50 possible landmark classes. The dropout layers in this section are to reduce the risk of overfitting by turning off some of the weights in each training step.

Initially I only used dropout on the fully connected layers at the end of my model, but I was getting accuracies below 5%. I then examined the dataset further and realised that ~5000 images split between 50 categories means only ~100 images per category, making this a relatively small dataset. Thus the risk of overfitting is quite high, and this ledd me to add in optional batch normalisation layers after the first few convolutional layers. This brought my accuracy up to 40%.

````python
def train(n_epochs, loaders, model, optimizer, criterion, use_cuda, save_path):
    """returns trained model"""
    # initialize tracker for minimum validation loss
    valid_loss_min = np.Inf
    #class_correct = list(0. for i in range(10))
    #class_total = list(0. for i in range(10))
    
    for epoch in range(1, n_epochs+1):
        # initialize variables to monitor training and validation loss
        train_loss = 0.0
        valid_loss = 0.0
        
        ###################
        # train the model #
        ###################
        # set the model to training mode
        model.train()
        for batch_idx, (data, target) in enumerate(loaders['train']):
            # move to GPU
            if use_cuda:
                data, target = data.cuda(), target.cuda()

            ## TODO: find the loss and update the model parameters accordingly
            ## record the average training loss, using something like
            ## train_loss = train_loss + ((1 / (batch_idx + 1)) * (loss.data.item() - train_loss))
            optimizer.zero_grad()
            
            output = model(data)
            
            loss = criterion(output, target)
            
            loss.backward()
            optimizer.step()

            train_loss += ((1 / (batch_idx + 1)) * (loss.data.item() - train_loss))
            
        ######################    
        # validate the model #
        ######################
        # set the model to evaluation mode
        model.eval()
        for batch_idx, (data, target) in enumerate(loaders['valid']):
            # move to GPU
            if use_cuda:
                data, target = data.cuda(), target.cuda()

            ## TODO: update average validation loss 

                output = model(data)
                
                loss = criterion(output, target)
            
                valid_loss += ((1 / (batch_idx + 1)) * (loss.data.item() - valid_loss))

        
        train_loss = train_loss / len(loaders['train'])
        valid_loss = valid_loss / len(loaders['valid'])
        
        # print training/validation statistics 
        print('Epoch: {} \tTraining Loss: {:.6f} \tValidation Loss: {:.10f}'.format(
            epoch, 
            train_loss,
            valid_loss
            ))

        ## TODO: if the validation loss has decreased, save the model at the filepath stored in save_path
                                      
        if valid_loss <= valid_loss_min:
            print('Validation loss decreased ({:.6f} --> {:.10f}).  Saving model ...'.format(
            valid_loss_min,
            valid_loss))
                              
            torch.save(model.state_dict(), save_path)
            valid_loss_min = valid_loss
        
    return model
````

````python
## TODO: Specify model architecture

# 1. Load the pretrained VGG-16 model from pytorch
model_transfer = models.vgg16(pretrained=True)
print(model_transfer)

# Find the number of parameters in the final fully connected layer
print(model_transfer.classifier[6].in_features)
print(model_transfer.classifier[6].out_features)

# 2. Freeze training for all "features" layers, so the net acts as a fixed feature extractor
for param in model_transfer.features.parameters():
    param.requires_grad = False
    
# 3. Remove the last linear layer and replace with our own one suited to our classification task
n_inputs = model_transfer.classifier[6].in_features

# add last linear layer (n_inputs -> 50 landmark classes)
# new layers automatically have requires_grad = True
last_layer = nn.Linear(n_inputs, 50)

model_transfer.classifier[6] = last_layer

# check to see that the last layer produces the expected number of outputs (50)
print("Final fully connected layer out_features: ", model_transfer.classifier[6].out_features)

#-#-# Do NOT modify the code below this line. #-#-#

if use_cuda:
    model_transfer = model_transfer.cuda()
    
````

__Answer:__  Our training dataset has ~5000 images in it, and with 50 classes this gives an average of ~100 images per class.
